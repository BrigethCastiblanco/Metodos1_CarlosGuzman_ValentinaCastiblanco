```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
p1 = np.linspace(0.1,0.9,100)
p2 = np.linspace(0.1,0.5,100)
```

```python
P1, P2 = np.meshgrid(p1,p2)
```

```python
res = np.empty_like(P1)
res = (P1*P2+2*(1-P1)*P2+2*P1*(1-P2)+(1-P1)*(1-P2))/4
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(2,1,1,projection='3d')
superficie = ax.plot_surface(P1,P2,res,cmap=cm.coolwarm)
ax.set_xlabel("P1",c='k',fontsize=15)
ax.set_ylabel("\n\n\nP2",c='k',fontsize=15)
ax.view_init(45,85)
```

```python
print('La probabilidad de minimo es', np.min(res))
print('La probabilidad de maxima es', np.max(res))
```
