```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
angulo=np.linspace(0,2*np.pi)
circulox=(2*np.cos(angulo))
circuloy=(1.5*np.sin(angulo))
```

```python
h=0.001
```

```python
X, Y= np.meshgrid(x,y)
fuera_circulo=(np.sqrt((X**2)+(Y**2))>=1.7)
X=X[fuera_circulo]
Y=Y[fuera_circulo]
```

```python
def potencial(x,y):
    return (2*x)*(1-((2**2)/((x**2)+(y**2))))
```

```python

def derivadax(f,x,y,h):
    return (f(x+h,y)-f(x-h,y))/(2*h)
def derivaday(f,x,y,h):
    return (f(x,y+h)-f(x,y-h))/(2*h)
```

```python
vx=derivadax(potencial,X,Y,h)
vy=derivaday(potencial,X,Y,h)
```

```python
fig1 = plt.figure(figsize=(5,5))
ax = fig1.add_subplot(1,1,1)
ax.plot(circulox,circuloy)
for i in range(len(X)):
    
        ax.quiver(X[i],Y[i],vx[i],vy[i],color='r',alpha=0.7)
```
