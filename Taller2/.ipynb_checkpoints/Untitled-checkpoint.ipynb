```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)

```

```python
h=0.001
```

```python
def potencial(x,y):
    
    return (2*x)*(1-((2**2)/((x**2)+(y**2))))
```

```python
X, Y= np.meshgrid(x,y)
phi=potencial(X,Y)
```

```python
def derivadax(f,x,y,h):
    return (f(x+h,y)-f(x-h,y))/(2*h)
def derivaday(f,x,y,h):
    return (f(x,y+h)-f(x,y-h))/(2*h)
```

```python
vx=np.zeros((25,25))
vy=np.zeros((25,25))

for i in range(25):
    for j in range(25):
        vx[i,j] = derivadax(potencial,x[i],y[j],h)
        vy[i,j] = -derivaday(potencial,y[i],y[j],h)

vy
```

```python
vx=vx.T
vy=vy.T


fig1 = plt.figure(figsize=(5,5))
ax = fig1.add_subplot(1,1,1)

for i in range(1,len(x)):
    for j in range(1,len(y)):
        ax.quiver(x[i],y[j],vx[i,j],vy[i,j],color='r')
```

```python



PartialOperator = (lambda f,x,y,h : (f(x+h,y)-f(x-h,y))/(2*h),\
           lambda f,x,y,h: (f(x,y+h)-f(x,y-h))/(2*h))
Ex = PartialOperator[0](potencial,X,Y,h)
Ey = -PartialOperator[1](potencial,X,Y,h)
Ex = Ex.T
Ey = Ey.T
fig1 = plt.figure(figsize=(5,5))
ax = fig1.add_subplot(1,1,1)

for i in range(1,len(x)):
    for j in range(1,len(y)):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j],color='r')
```
