```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
from tqdm import tqdm
```

```python
# Definamos el sistema en una tupla
G=(lambda x1,x2,x3,x4:(x1**2 + x2**2 + x3**2 + x4**2 -1))
```

```python
def GetVectorF(G,r):
    
    dim = 1
    
    v = np.zeros(dim)
    
    
    v[0] = G(r[0],r[1],r[2],r[3])
        
    return v
```

```python
GetVectorF(G,[1,0,0,1])
```

```python
def GetJacobian(G,r,h=1e-6):
    
    
    
    J = np.zeros((1,4))
    
   
    J[0,0] = (  G(r[0]+h,r[1],r[2],r[3]) - G(r[0]-h,r[1],r[2],r[3]) )/(2*h)
    J[0,1] = (  G(r[0],r[1]+h,r[2],r[3]) - G(r[0],r[1]-h,r[2],r[3]) )/(2*h)
    J[0,2] = (  G(r[0],r[1],r[2]+h,r[3]) - G(r[0],r[1],r[2]-h,r[3]) )/(2*h)
    J[0,3] = (  G(r[0],r[1],r[2],r[3]+h) - G(r[0],r[1],r[2],r[3]-h) )/(2*h)
        
    return J.T
```

```python
def GetMetric(G,r):
    v = GetVectorF(G,r)
    return 0.5*np.linalg.norm(v)**2
```

```python
GetMetric(G,[1,0,0,1])
```

```python
def GetFig(F,R,it):
    
    fig = plt.figure(figsize=(8,4))
    
    labels = ['X1','X2','X3','X4']
    
    ax = fig.add_subplot(1,2,1)
    ax1 = fig.add_subplot(1,2,2)

    ax.set_title('Metric: %.20f' %(F[it]))

    ax.plot(F[:it])
    ax.set_xlabel('%.0f' %(it))
    ax.set_yscale('log')
    ax1.plot(R[:it],label=labels)
    ax1.set_xlabel('%.0f' %(it))
    ax1.legend(loc=0)
    
    plt.show()
```

```python
def GetSolve(G,r,lr=1e-3,epochs=int(1e5),error=1e-7,visualize=False):
    
    d = 1
    it = 0
    Vector_F = np.array([])
    
    R_vector = np.array(r)
    
    while d > error and it < epochs:
        
        CurrentF = GetMetric(G,r)
        
        J = GetJacobian(G,r)
        
        GVector = GetVectorF(G,r)
        
        #Machine Learning
        r -= lr*np.dot(J,GVector) 
        
        R_vector = np.vstack((R_vector,r))
        
        NewF = GetMetric(G,r)
        
        
        Vector_F = np.append(Vector_F,NewF)
        if NewF !=0:
        
            d = np.abs( CurrentF - NewF )/NewF
        else:
            d = 0
        
        if it%500 == 0 and visualize:
            
            #print(it,d)
            clear_output(wait=True)
            GetFig(Vector_F,R_vector,it)
            time.sleep(0.01)
            
        it += 1
        
    if d < error and visualize:
        print(' Entrenamiento completo ', d, 'iteraciones', it)
        
    if it == epochs and visualize:
        print(' Entrenamiento no completado ')
        
    return r,it,Vector_F,R_vector
```

```python
xsol,it,F,R = GetSolve(G,np.array([1,-1,0,-0.2]),lr=1e-3,visualize=True)
```

```python
xsol,it,F,R = GetSolve(G,np.array([0,0,0,0.1]),lr=1e-3)
```

```python
N = 1000 

sphere = np.zeros([N,4])

for i in tqdm(range(N)):
    
    randp = np.random.uniform(-1,1,4)
    sol1,sol2,sol3,sol4 = GetSolve(G,randp, lr=1e-3)
    sphere[i,:] = sol1
```

```python
metrics = np.zeros(N)

for i in range(N):
    metrics [i] = GetMetric(G,sphere[i,:] )
```
