```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

n=5
delta=[0, np.pi/4, np.pi/2]
teta=np.linspace(0,2*np.pi,300)
nx=[1,2,3,4,5]
ny=[1,4,5,6,7]

```

```python
def desfase1(nx:list,ny:list):
    X1=[]
    Y1=[]
    k=1
    for i in range(n):
        for j in range(n):
         x=np.sin(nx[i]*teta)
        
         y=np.sin(ny[j]*teta)
         fig=plt.subplot(n,n,k)
         k+=1
         fig.plot(x,y)
         plt.axis('off')
```

```python
def desfase2(nx:list,ny:list):
    X1=[]
    Y1=[]
    k=1
    for i in range(n):
        for j in range(n):
         x=np.sin(nx[i]*teta)
        
         y=np.sin(ny[j]*teta + delta[1])
         fig=plt.subplot(n,n,k)
         k+=1
         fig.plot(x,y)
         plt.axis('off')
```

```python
def desfase3(nx:list,ny:list):
    X1=[]
    Y1=[]
    k=1
    for i in range(n):
        for j in range(n):
         x=np.sin(nx[i]*teta)
        
         y=np.sin(ny[j]*teta + delta[2])
         fig=plt.subplot(n,n,k)
         k+=1
         fig.plot(x,y)
         plt.axis('off')
            
```

```python
print('El primer desfase para cuando delta es 0 es: ')
(desfase1(nx,ny))
```

```python
print('El segundo desfase para cuando delta es pi/4 es: ')
(desfase2(nx,ny))
```

```python
print('El tercer desfase para cuando delta es pi/2 es: ')
(desfase3(nx,ny))
```
