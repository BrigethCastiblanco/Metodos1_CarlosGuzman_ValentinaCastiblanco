```python
import matplotlib.pyplot as plt
ejex=list(range(1,1001))
lista=list(range(2,500000))
def primos(x:int)->bool:
    res=True
    for f in range(2,x):
        if x%f ==0 :
            res=False
    return res
def primos1(x:list)->list:
    
    lista=[]
    
    i=0
    while (i<=len(x)) and (len(lista)<=999):
        
        v=x[i]
        
        if primos(v)==True:
            lista.append(v)
        
        i+=1
        
    return lista
```

```python
j=primos1(lista)
print('Los primeros 1000 numeros primos son: \n'+ str(j))
```

```python
def diezprim(lista:list)->list:
    x=[]
    i=0
    while i<=9:
        v=lista[i]
        
        x.append(v)
        
        i+=1
    return x
    
```

```python
l=diezprim(j)
print('Los diez primeros numeros primos son: \n' + str(l))
```
