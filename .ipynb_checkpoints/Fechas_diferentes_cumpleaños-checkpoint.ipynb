```python
import matplotlib.pyplot as plt
```

$$ P = \displaystyle\frac{365-n_{i}}{365^{n}}$$

```python
no_birthday=[]
n = 365
year_days = 365
p=1
for i in range(n):
    p_no_match = ((year_days-i)/year_days)
    print(i, p_no_match)
    no_birthday.append(p_no_match)
```

```python
prob = no_birthday[:79]
```

```python
import numpy as np
y = prob
x = np.linspace(0,79,79)
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(2,2,1)

ax.scatter(x,y)
plt.xlabel("n")
plt.ylabel("probabilidad")
```
